```python
import os
import requests
import pandas as pd
from io import StringIO


# Retrieve the value of the environment variable
api_key = os.environ.get('API_KEY')
```

```python
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=XLK&outputsize=compact&apikey={api_key}'

r = requests.get(url)
data = r.json()

print(data)
```

```python
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=XLK&outputsize=compact&datatype=csv&apikey={api_key}'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the CSV content to a pandas DataFrame
    csv_data = response.text
    df = pd.read_csv(StringIO(csv_data))

    # Display the DataFrame
    print(df)
else:
    print("Failed to fetch data. Status code:", response.status_code)
```
